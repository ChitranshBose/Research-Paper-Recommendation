```python
import numpy as np 
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('D:/research_new_dataset.csv')
df.head()
```

```python
df.info
```

```python
n=len(df)
print(n)
l=[]
for i in range(n):
    l.append(i)
#print(l)
ll=pd.DataFrame(l)
ll.head()
```

```python
df['Index']=ll[0]
df.head()
```

```python
x=list(df.loc[2])
print(x)
```

```python
def get_index_from_title(title):
    return df[df.TITLE == title]["Index"].values

def get_title_from_index(index):
    return df[df.index == index]["TITLE"].values[0]
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["TITLE"])
cosine_sim = cosine_similarity(count_matrix)
paper_user_likes = "Reliability of Computational Experiments on Virtualised Hardware"
```

```python
paper_index = get_index_from_title(paper_user_likes)
similar_papers = list(enumerate(cosine_sim[paper_index]))
sorted_similar_papers = sorted(similar_papers,key = lambda x:x[1],reverse = True)
print(sorted_similar_papers)
```

```python
i = 0
for element in sorted_similar_papers:
    print(get_title_from_index(element[0]))
    i = i+1
    if i>10:
        break
```
