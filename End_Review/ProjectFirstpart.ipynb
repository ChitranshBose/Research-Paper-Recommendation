```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

query = "calculation of prompt diphoton production"

stemmed_query = []
for word in query.split():
    stemmed_word = ps.stem(word)
    stemmed_query.append(stemmed_word)

query_vector = cv.transform([' '.join(stemmed_query)]).toarray()

print("Stemmed query: ", ' '.join(stemmed_query))
print("Query vector: ", query_vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from termcolor import colored

cos_similarities = cosine_similarity(query_vector, vectors_tag)
top_k = 5
similar_indices = cos_similarities.argsort()[0][-top_k:]
print("Top {} most similar titles:".format(top_k))
print("******************************************************************************")
a=0
for i in similar_indices:
    a=a+1
    text = 'RECOMMENDATION------------------'
    big_red_text = colored(text, 'red', attrs=['bold'])
    print(big_red_text,a)
    print(paper['title'][i],"   ")
    id=paper['id'][i]
    print('url = 'f"https://arxiv.org/pdf/{id}")
     

```

```python
#EVALUATION USING  1 gram 
from nltk.translate.bleu_score import sentence_bleu
print("-----------------EVALATION USING BLUE SCORE -----------------------------")

for i in similar_indices:
    title = paper['title'][i]
    score = sentence_bleu([stemmed_query], title.split(), weights=(1, 0, 0, 0))
    print("--------------------------------------------------")
    print("{}: {:.6f}".format(title, score))
```

```python
# EVALUATION USING 2 gram
from nltk.translate.bleu_score import sentence_bleu

print("-----------------EVALUATION USING BLUE SCORE (2-gram) -----------------------------")

for i in similar_indices:
    title = paper['title'][i]
    score = sentence_bleu([stemmed_query], title.split(), weights=(0.5, 0.5, 0, 0))
    print("--------------------------------------------------")
    print("{}: {:.6f}".format(title, score))
```
