```python
import sklearn
import bs4
import pickle as pk
import numpy as np
import pandas as pd
import nltk
import docx as doc
from pathlib import Path
import shutil
import os
import re
from bs4 import BeautifulSoup
import nltk
nltk.download('punkt')
from nltk import word_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk import FreqDist
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize, pos_tag
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4') 
```

```python
data = doc.Document('Detection of abusive comments.docx').paragraphs
```

```python
lst = []
for x in data:
  lst.append(x.text.lower())
print(lst)
```

```python
def tokenize_sent(data):
    token = nltk.word_tokenize(data)
    return [ word for word in token if len(word)>=3 and word.isalpha()]
```

```python
df = pd.DataFrame(lst)
df
```

```python
df['output'] = df.apply(lambda x: tokenize_sent(x[0]),axis=1)
df
```

```python
stopwords = stopwords.words('english')
#stop_words = set(stopwords.words("english"))
def removeWords(data):
  
  output = []
  for word in data:
    if word not in stopwords:
      output.append(word)
  return output
```

```python
df['output'] = df['output'].apply(removeWords)
df
```

```python
def lemma_sent(data):
  output = []
  lst = ['a','r','n','v']
  wordNetLemmatizer = WordNetLemmatizer()
  posTag = pos_tag(data)
  for word, tag in posTag:
    pos = tag[0].lower()

    if pos not in lst:
      pos = 'n'
    output.append(wordNetLemmatizer.lemmatize(word,pos))
  return output
```

```python
#df['output'] = df.apply(lambda x: lemma_sent(x[0]),axis=1)
df['output'] = df['output'].apply(lemma_sent)
df
```
